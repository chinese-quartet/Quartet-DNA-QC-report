```python
with open('./multiqc_data/multiqc_happy_snp_data.json') as hap_snp_json:
		happy_snp = json.load(hap_snp_json)

dat_snp = pd.DataFrame.from_records(happy_snp)

dat_snp = dat_snp.loc[:, dat_snp.columns.str.endswith('ALL')]
dat_transposed_snp = dat_snp.T
dat_transposed_snp = dat_transposed_snp.loc[:,['sample_id','QUERY.TOTAL_snp','METRIC.Precision_snp','METRIC.Recall_snp']]
snp = dat_transposed_snp[['SNP' in s for s in dat_transposed_snp.index]]
```

```python
indel.reset_index(drop=True, inplace=True)
```

```python
snp.reset_index(drop=True, inplace=True)
```

```python
benchmark = pd.concat([snp, indel], axis=1)
```

```python
benchmark.columns
```

```python
benchmark = benchmark[['sample_id', 'QUERY.TOTAL_snp', 'QUERY.TOTAL_indel','METRIC.Precision_snp',
       'METRIC.Precision_indel','METRIC.Recall_snp', 
        'METRIC.Recall_indel']]
benchmark.columns = ['Sample','sample_id','SNV number','INDEL number','SNV precision','INDEL precision','SNV recall','INDEL recall']
benchmark = benchmark[['Sample','SNV number','INDEL number','SNV precision','INDEL precision','SNV recall','INDEL recall']]
```

```python
benchmark['SNV precision'] = benchmark['SNV precision'].astype(float)
benchmark['INDEL precision'] = benchmark['INDEL precision'].astype(float)
benchmark['SNV recall'] = benchmark['SNV recall'].astype(float)
benchmark['INDEL recall'] = benchmark['INDEL recall'].astype(float)
benchmark['SNV precision'] = benchmark['SNV precision'] * 100
benchmark['INDEL precision'] = benchmark['INDEL precision'] * 100
benchmark['SNV recall'] = benchmark['SNV recall'] * 100
benchmark['INDEL recall'] = benchmark['INDEL recall']* 100
```

```python
benchmark = benchmark.round(2)
benchmark.to_csv('./test_multiqc/variants.calling.qc.txt',sep="\t",index=0)
```
